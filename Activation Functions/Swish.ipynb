```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import backend as K
```

```python
def swish_function(x):
    return x * sigmoid_function(x)

def swish_function_derivative(x):
    sig = sigmoid_function(x)
    return sig + x * sig * (1 - sig)
```

```python
def plot_activation_and_derivative(x, activation_func, derivative_func, title):
    y_values = activation_func(x)
    y_derivative_values = derivative_func(x)

    plt.figure(figsize=(12, 6))

    # Plot the activation function
    plt.subplot(1, 2, 1)
    plt.plot(x, y_values, label=f'{title} Activation')
    plt.title(f'{title} Activation Function')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.legend()

    # Plot the derivative of the activation function
    plt.subplot(1, 2, 2)
    plt.plot(x, y_derivative_values, label=f'{title} Derivative', color='orange')
    plt.title(f'{title} Derivative Function')
    plt.xlabel('x')
    plt.ylabel('f\'(x)')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
x_values = np.linspace(-10, 10, 400)
```

```python
plot_activation_and_derivative(x_values, swish_function, swish_function_derivative, 'Swish')
```
