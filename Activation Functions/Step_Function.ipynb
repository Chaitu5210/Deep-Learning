```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import backend as K
```

```python
def step_function(x):
    return np.where(x >= 0, 1, 0)

def step_function_derivative(x):
    return np.zeros_like(x)  # The derivative of the step function is zero almost everywhere
```

```python
def plot_activation_and_derivative(x, activation_func, derivative_func, title):
    y_values = activation_func(x)
    y_derivative_values = derivative_func(x)

    plt.figure(figsize=(12, 6))

    # Plot the activation function
    plt.subplot(1, 2, 1)
    plt.plot(x, y_values, label=f'{title} Activation')
    plt.title(f'{title} Activation Function')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.legend()

    # Plot the derivative of the activation function
    plt.subplot(1, 2, 2)
    plt.plot(x, y_derivative_values, label=f'{title} Derivative', color='orange')
    plt.title(f'{title} Derivative Function')
    plt.xlabel('x')
    plt.ylabel('f\'(x)')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
x_values = np.linspace(-10, 10, 400)
```

```python
# Plot Step Function
plot_activation_and_derivative(x_values, step_function, step_function_derivative, 'Step')
```
